Converts floating point number to fixed point.

```python
def to_fixed(f,e):
    a = f* (2**e)
    b = int(round(a))
    #print(b)
    if a < 0:
        # next three lines turns b into it's 2's complement.
        b = abs(b)
        b = ~b
        b = b + 1
    return b
```

Converts fixed point number to float.

```python
def to_float(x,e):
    c = abs(x)
    sign = 1 
    if x < 0:
        # convert back from two's complement
        c = x - 1 
        c = ~c
        sign = -1
    f = (1.0 * c) / (2 ** e)
    f = f * sign
    return f
```

Converts positive or negative integer to two's complement representation hexadecimal.

```python
def tohex(val, nbits):
    return hex((val + (1 << nbits)) % (1 << nbits))

print(tohex(11128, 32).lstrip('0x').upper())
```

Converts the contents of a .csv file that contains floating point numbers, all to fixed point, and then converts those integer values to hexadecimal, and finally outputs the result on screen.

```python
import csv

with open('C:\\Users\\Marko Visnjic\\steps\\4.csv', newline='') as csvfile:
    data = list(csv.reader(csvfile))
    
array_of_values = [];
array_of_fixed = [];
for _ in data:
    for i in _:
        temp = i.split('e')
        temp[0] = float(temp[0])
        temp[1] = float(temp[1])
        
        value = to_fixed(temp[0]*(pow(10,temp[1])),15)
        
        array_of_fixed.append(value);
        
        hex_value = tohex(value,32).replace('0x','')
        
        if (len(hex_value) == 0):
            hex_value = '00000000' + hex_value
            
        elif (len(hex_value) == 1):
            hex_value = '0000000' + hex_value
            
        elif (len(hex_value) == 2):
            hex_value = '000000' + hex_value
            
        elif (len(hex_value) == 3):
            hex_value = '00000' + hex_value
            
        elif (len(hex_value) == 4):
            hex_value = '0000' + hex_value
        
        elif (len(hex_value) == 5):
            hex_value = '000' + hex_value
        
        elif (len(hex_value) == 6):
            hex_value = '00' + hex_value
            
        elif (len(hex_value) == 7):
            hex_value = '0' + hex_value
            
        print(hex_value)
        array_of_values.append(hex_value);
```
