Normalize 0 to 255 input pixel map to value in between 0 and 1, so that we convert integer representation to float representation.

```python
import csv
import numpy as np

with open('five_prenormalize.csv', newline='') as csvfile:
    data = list(csv.reader(csvfile))
    
pixelmap_values_infloat = []
    
for _ in data:
    for i in _:
        temp = i.split('+')
        temp[0] = temp[0].replace('e', '')
        temp[0] = float(temp[0])
        temp[1] = float(temp[1])

        value = int(round(temp[0]*(pow(10,temp[1]))))
        
        #perform linear mapping to convert 0 to 255 range to usable 0 to 1 float range for NN
        x = value
        y = x / 255.0
            
        print(y)
        pixelmap_values_infloat.append(y)
```
